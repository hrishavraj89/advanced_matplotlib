```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

COLORED SCATTERPLOTS

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
#assign number to the recurrent column(here species)
iris['Species'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2})
```

```python
#first we have to assign recurrent values a number, the reassign that column in the actual dataset and the use the hidden parameter 'c' like this.
#we can change the color styles using 'cmaps' 
#colorbar helps use to understand the which color represent which values.
#alpha helps to control the opacity of the scatter points.
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
#plot size / Figure size
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```
