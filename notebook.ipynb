```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

COLORED SCATTERPLOTS

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
#assign number to the recurrent column(here species)
iris['Species'] = iris['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2})
```

```python
#first we have to assign recurrent values a number, the reassign that column in the actual dataset and the use the hidden parameter 'c' like this.
#we can change the color styles using 'cmaps' 
#colorbar helps use to understand the which color represent which values.
#alpha helps to control the opacity of the scatter points.
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
#plot size / Figure size
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c = iris['Species'], cmap = 'jet', alpha = 0.6)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
plt.show()
```

```python
batters = pd.read_csv('batter.csv')
batters
```

```python
batters.shape
```

```python
#randomly sselecting 25 batsman from top 100
sample_df = batters.head(100).sample(25, random_state = 5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'])
```

```python
#annotations on a dataframe
#here we run a loop to plot all the annotations

plt.figure(figsize = (15,7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])   #'s' is used to define the size of the points depending on how much runs scored
for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

PLOTTING HORIZONTAL AND VERTICAL LINES

```python
plt.figure(figsize = (15,7))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s = sample_df['runs'])   #'s' is used to define the size of the points depending on how much runs scored

plt.axhline(130, color = 'red')
plt.axhline(140, color = 'green')
plt.axvline(30, color = 'red')

for i in range (sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

SUBPLOTS

```python
#normal approach
plt.figure(figsize=(15,7))
plt.scatter(batters['avg'], batters['strike_rate'], marker = '+', color = 'red')
plt.title('Average vs Strike Rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

```python
#another approach
fig, ax = plt.subplots(figsize=(15,7))
ax.scatter(batters['avg'], batters['strike_rate'], marker = '+', color = 'red')
ax.set_title('Average vs Strike Rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')
fig.show()
```

```python
#nrows = no. of rows
#ncols = no. of colums\
# nrows*ncols = total number of graphs you wanna create
# 'sharex=True' help to share the label across all the graphs
fig ,ax = plt.subplots(nrows = 2, ncols = 1, sharex=True, figsize=(10,6))

ax[0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[0].set_title('Average vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batters['avg'], batters['runs'])
ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
```

3-D SURFACE PLOTS
- used to determine the working of the model

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
xx.shape
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(x,yy,z, cmap='viridis')
fig.colorbar(p)   #this is how we use colorbar in these.
```

```python
z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(x,yy,z, cmap='viridis')
fig.colorbar(p)
```
